```python
from urllib.request import urlopen 
from bs4 import BeautifulSoup 
import re
import json
import csv
import urllib.error
```

```python
class Letters: #my class 
    
    def __init__(self, url): #url is my one attrubute of the my class
        self.url = url
    
              
    def retrieve_shops_first_letter(self): #one function
        
    
        """Get all letter links from the main page"""
        base_url = 'https://www.mystore411.com'
        html = urlopen(self.url)
        bs = BeautifulSoup(html, 'html.parser')
        urls = []
        
        for link in bs.find('div', {'id':'breadcrumb'}).find_all('a', href=re.compile('^(/store/)((?!:).)*$')):
            if 'href' in link.attrs:
                letters = link.attrs['href']
                urls.append(letters)
                new_urls = ['{}{}'.format(base_url,i) for i in urls]
                
        return new_urls 
```

```python
object1 = Letters('https://www.mystore411.com') # base is my object containing attribute 'https://www.mystore411.com' as url 
object1.retrieve_shops_first_letter()
```

```python
class Stores:
    def __init__(self, newurl):
        self.newurl = newurl
        
    def retrieve_stores(self):
        """Get all links of all shops"""
        base_url = 'https://www.mystore411.com'
        stores = []
        for each in self.newurl:
            html = urlopen(each)
            soup = BeautifulSoup(html, "lxml" )
            #newnewurl = []
            for ul in soup.find_all('ol', class_='col2 d_store_list'):
                for li in ul.find_all('li'):
                    a = li.find('a')
                    newnewurl = a['href']
                    stores.append(newnewurl)
                    all_stores_links = ['{}{}'.format(base_url,i) for i in stores]
                    #print(all_stores_links)
        return(all_stores_links)
        
    
```

```python
object2 = Stores(object1.retrieve_shops_first_letter())
object2.retrieve_stores()
```
